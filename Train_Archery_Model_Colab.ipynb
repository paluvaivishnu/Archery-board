```python
from roboflow import Roboflow

# TODO: Paste your Roboflow API key below inside the quotes
rf = Roboflow(api_key="YOUR_API_KEY_HERE")

project = rf.workspace("vishnus-workspace-wetuu").project("bds-8hkd8")
version = project.version(1)
dataset = version.download("yolov11")
```

```python
from ultralytics import YOLO

# Load a fresh YOLO11m model
model = YOLO('yolo11m.pt')

# Train for 100 epochs (Crucial for high confidence!)
results = model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    project="archery_yolo",
    name="run1"
)
```

```python
# Export the trained model to ONNX
success = model.export(format='onnx', imgsz=640, simplify=True, opset=17)
print("Model successfully exported to:", success)
```
